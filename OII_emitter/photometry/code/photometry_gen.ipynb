# Derive SFR from VLA 3GHz

```python
def luminosity_1_4GHz(flux_3GHz, z, luminosity_distance, alpha=-0.7):
    '''
    flux: microJy
    luminosity_distance: Mpc
    return: W/Hz, rest-frame 1.4GHz luminosity
    '''
    return 4*np.pi*(luminosity_distance*u.Mpc.to(u.m))**2*(flux_3GHz*1e-32)*(1.4/3)**alpha/(1+z)**(1+alpha)
def q_TIR(luminosity_1_4GHz):
    '''
    the ratio of infrared-to-radio luminosity
    '''
    return -0.153*np.log10(luminosity_1_4GHz)+5.9
def L_TIR(q_TIR, luminosity_1_4GHz):
    '''
    luminosity_1_4GHz: W/Hz
    return: W
    '''
    return 10**q_TIR*luminosity_1_4GHz*3.75*10**12
def SFR_TIR(L_TIR, log_C=43.41):
    '''
    L_TIR: W
    return: Msun/yr
    '''
    log_SFR = np.log10(L_TIR*10**7)-log_C
    return 10**log_SFR
```

```python
flux_3GHz = 24.6
lum_1_4GHz = luminosity_1_4GHz(flux_3GHz, 0.924, D_lum.value)
lum_1_4GHz
```

```python
q_TIR_ = q_TIR(lum_1_4GHz)
q_TIR_
```

```python
L_TIR = L_TIR(q_TIR_, lum_1_4GHz)
L_TIR
```

```python
SFR_TIR = SFR_TIR(L_TIR)
SFR_TIR
```

```python
print('L_TIR:', round(L_TIR, 2), 'W',
      '\nq_TIR:', round(q_TIR, 2),
      '\nlum_1_4GHz:', round(lum_1_4GHz, 2), 'W/Hz',
      '\nSFR_TIR:', round(SFR_TIR, 2), 'Msun/yr',)
```

```python
from scipy.stats import norm
num = 0
plt.figure(figsize=(32,15))
for telescope_band in phot.image_files.keys():
    if 'SPITZER' in telescope_band or 'GEMINI' in telescope_band or 'VISTA' in telescope_band: continue
    num += 1
    data = phot.image_files[telescope_band]
    bkg_map, rms_map, source_mask = phot.load_background(telescope_band)
    pure_bkgsub = np.ma.array(data-bkg_map, mask=source_mask)
    pure_bkgsub_cutout = Cutout2D(pure_bkgsub, source_position, size=10*u.arcsec, wcs=phot.wcss[telescope_band])
    pure_bkgsub_cutout_data1D = pure_bkgsub_cutout.data.data.flatten()[~pure_bkgsub_cutout.data.mask.flatten()]
    mu, std = norm.fit(pure_bkgsub_cutout_data1D)
    nm = ImageNormalize(pure_bkgsub_cutout.data, interval=ZScaleInterval())
    plt.subplot(5, 10, num*2-1)
    plt.imshow(pure_bkgsub_cutout.data, origin='lower', norm=nm)
    plt.subplot(5, 10, num*2)
    plt.hist(pure_bkgsub.flatten(), bins=100, edgecolor='red', histtype='step', density=True)
    plt.hist(pure_bkgsub_cutout_data1D, bins=30, edgecolor='blue', histtype='step', density=True)
    xlim0 = np.abs([np.percentile(pure_bkgsub_cutout_data1D, 0.1), np.percentile(pure_bkgsub_cutout_data1D, 99.9)]).max()
    plt.xlim(-xlim0, xlim0)
    ymin, ymax = plt.ylim()
    x = np.linspace(-xlim0, xlim0, 10000) 
    p = norm.pdf(x, mu, std) 
    plt.plot(x, p, 'k') 
    plt.vlines([np.median(pure_bkgsub_cutout_data1D)], [ymin], [ymax], color='blue', linestyle='dashed')
    plt.vlines([np.median(pure_bkgsub.flatten())], [ymin], [ymax], color='red', linestyle='dashed')
    plt.vlines([np.percentile(pure_bkgsub_cutout_data1D, 16), np.percentile(pure_bkgsub_cutout_data1D, 84)], ymin, ymax, color='red', linestyle='dotted') # 1 sigma
    plt.fill_between(x, p, where=((x>(mu-std))&(x<(mu+std))), facecolor='k', alpha=0.2)
    plt.title(telescope_band)
plt.show()
```
