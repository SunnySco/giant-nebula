```python
%matplotlib inline
import matplotlib.pyplot as plt
import pyneb as pn
```

Photoionized gas usually has a temperature around 10^4 K, while shocked gas and turbulent mixing has higher temperatures, approaching 10^5 K.

```python
O3_EG = pn.EmisGrid(elem='O', spec=3, n_tem=100, n_den=100, 
                    tem_min=1e5, tem_max=1e6, den_min=10., 
                    den_max=1.e8, restore_file=None, atomObj=None)
```

```python
f, ax = plt.subplots(figsize=(7,5))
O3_EG.plotContours(to_eval = 'L(5007)/L(4363)', ax=ax, log_levels=False)
ax.set_yscale('log')
f.savefig('OIII_diag.pdf')
```
